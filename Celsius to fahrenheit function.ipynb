```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
met_data = pd.read_csv("FC_Ag_5min_2023-12-01_2023-12-31.csv", parse_dates=["Date and Time"])

met_data.head() # show the first five rows
```

```python
def cf(temp):
    """
    Convert Fahrenheit temperatures to Celsius.

    Parameters:
    temp: A list, Series, or NumPy array of temperatures in Fahrenheit.

    Returns:
    NumPy array: Temperatures converted to Celsius, rounded to 2 decimal places.
    
    """
    
    temp=np.array(temp)
    
    ctemp=np.round((temp-32)*(5/9),2)
    
    return ctemp  

met_data["Air Temp(C)"]=cf(met_data["Air Temp"])

met_data["Dewpoint(C)"]=cf(met_data["Dewpoint"])

plt.plot(met_data["Date and Time"],met_data["Air Temp(C)"], color="red", label="Air Temperature(C)")

plt.plot(met_data["Date and Time"],met_data["Dewpoint(C)"], color="blue", label="Dewpoint Temperature(C)")

plt.xlabel("Time (days)", fontsize=10, fontweight="bold")

plt.ylabel("Temperature (C)", fontsize=10, fontweight="bold")

plt.xticks(rotation=-45)

plt.title("Air and Dewpoint Temperatures over December 2022", fontsize=12, fontweight="bold")

plt.margins(x=0)

plt.legend()

plt.show()
```
